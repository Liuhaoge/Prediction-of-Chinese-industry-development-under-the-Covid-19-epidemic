```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
import matplotlib.pyplot as plt
```

```python
import tushare as ts
```

```python
pro = ts.pro_api('22f7870869eb77d0aa2206f2d7559586532133b97ca962dfcf9e5611')
```

```python
# 所有股票列表
data = pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
data # 4443个
```

```python
data.ts_code[2300]
```

### 股价元数据

```python
df = pro.daily(ts_code=data.ts_code[0], start_date='20200122', end_date='20200619', fields='trade_date,pre_close,open,high,low,vol,amount,change')
df['trend']=df.change.apply(lambda x: 0 if x<0 else 1)
df=df.drop(['change'],axis=1)

df
```

```python
for i in range(3000,3500):
    tmp = pro.daily(ts_code=data.ts_code[i], start_date='20200831', end_date='20210820', fields='trade_date,pre_close,open,high,low,vol,amount,change')
    tmp['trend']=tmp.change.apply(lambda x: 0 if x<0 else 1)
    tmp=tmp.drop(['change'],axis=1)
    df=df.append(tmp)

df.size
```

```python
df=df.reset_index(drop=True)
```

```python
# df.to_csv('raw_stock_data.csv',columns=['trade_date','open','high','low','pre_close','vol','amount','trend'])
```

### 1.中国疫情数据


```python
Covid_China = pro.ncov_global(country='中国',fields='area_id,publish_date,country_enname,confirmed_num_now,confirmed_num,cured_num,dead_num')
```

```python
Covid_China=Covid_China.groupby('publish_date').sum()
```

```python
Covid_China.columns=['China_confirmed_num','China_confirmed_num_now','China_cured_num','China_dead_num']
```

```python
result1=pd.merge(df, Covid_China,left_on='trade_date', right_on='publish_date')
result1
```

### 2.世界疫情数据


```python
Covid_World = pro.ncov_global(start_date='20200122',fields='area_id,publish_date,country_enname,confirmed_num_now,confirmed_num,cured_num,dead_num')
```

```python
Covid_World = Covid_World.groupby('publish_date').sum()
```

```python
Covid_World.columns=['World_confirmed_num','World_confirmed_num_now','World_cured_num','World_dead_num']
```

```python
result2=pd.merge(df, Covid_World, left_on='trade_date', right_on='publish_date')
result2
```

### 3. 国内宏观数据

```python
# Ex-factory price index of industrial producers
industrial_price_index=pro.cn_ppi(start_m='202001', end_m='202006', fields='month,ppi_yoy,ppi_mom,ppi_accu')
consumer_price_index=pro.cn_cpi(start_m='202001', end_m='202006', fields='month,nt_val,nt_yoy,nt_mom,town_val,town_yoy,town_mom,cnt_val,cnt_yoy,cnt_mom')

China_macro_economy=pd.merge(industrial_price_index, consumer_price_index, left_on='month', right_on='month')
China_macro_economy
```

```python
df_tmp=df
df_tmp['trade_month']=df_tmp.trade_date.apply(lambda x: x[0:6])
df_tmp=df_tmp.drop('trade_date',axis=1)
df_tmp.head()
```

```python
result3=pd.merge(df_tmp, China_macro_economy, left_on='trade_month', right_on='month').drop('month',axis=1)
result3
```

### 数据的混合

```python
# 两个疫情数据的混合
result4=pd.merge(result1, Covid_World, left_on='trade_date', right_on='publish_date')
result4
```

```python
# 国外疫情与国内宏观的混合
result2_tmp=result2
result2_tmp['trade_month']=result2_tmp.trade_date.apply(lambda x: x[0:6])
result2_tmp=result2_tmp.drop('trade_date',axis=1)

result5=pd.merge(result2_tmp, China_macro_economy, left_on='trade_month', right_on='month').drop('month',axis=1)
result5
```

```python
# 国内疫情与国内宏观的混合
result1_tmp=result1
result1_tmp['trade_month']=result1_tmp.trade_date.apply(lambda x: x[0:6])
result1_tmp=result1_tmp.drop('trade_date',axis=1)

result6=pd.merge(result1_tmp, China_macro_economy, left_on='trade_month', right_on='month').drop('month',axis=1)
result6
```

```python
# 国内外疫情与国内宏观的混合，三类数据特征
result4_tmp=result4
result4_tmp['trade_month']=result4_tmp.trade_date.apply(lambda x: x[0:6])
result4_tmp=result4_tmp.drop('trade_date',axis=1)

result7=pd.merge(result4_tmp, China_macro_economy, left_on='trade_month', right_on='month').drop('month',axis=1)
result7
```
